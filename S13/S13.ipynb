```python
import LightningModel
from pytorch_lightning import LightningModule

myLitmodel = LightningModel.LitYolo()
```

```python
from pytorch_lightning import Trainer

trainer = Trainer(
    accelerator = 'gpu',
    max_epochs = 40
)

# Fit model
trainer.fit(myLitmodel) #train_loader,test_loader
```
